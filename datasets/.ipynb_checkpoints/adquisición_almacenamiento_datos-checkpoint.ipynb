### 1. Lee sobre las siguientes t√©cnicas de adquisici√≥n de datos: ###
- **APIs:**
Son interfaces que permiten interactuar con servicios externos para obtener datos.  A trav√©s de una API, puedes enviar solicitudes a un servidor externo para obtener informaci√≥n en tiempo real, como los datos del clima de OpenWeather o las publicaciones de Twitter. **C√≥mo interactuar con ellas:** Para interactuar con una API, generalmente se realiza una solicitud HTTP, por ejemplo, usando el m√©todo GET para obtener datos. A menudo, la API devuelve los datos en formato JSON o XML. Con Python, se puede usar bibliotecas como requests para hacer estas solicitudes y json para manipular los datos que se reciben.


- **Web Scraping:** Esta t√©cnica consiste en extraer informaci√≥n directamente desde sitios web. A diferencia de las APIs, que proporcionan datos estructurados, el web scraping extrae contenido directamente de las p√°ginas web, lo que puede requerir la navegaci√≥n y la interpretaci√≥n del HTML. Herramientas como BeautifulSoup o Scrapy se utilizan com√∫nmente en Python para realizar este tipo de tareas.

  
- **Bases de Datos:** Son sistemas de almacenamiento estructurado que organizan datos para su acceso, manipulaci√≥n y actualizaci√≥n. Python puede conectarse a bases de datos usando bibliotecas como sqlite3 (para bases de datos SQLite) o SQLAlchemy (para bases de datos m√°s complejas). Estas herramientas permiten realizar consultas SQL y obtener resultados en forma de tablas o conjuntos de datos.



### 2. Responde las siguientes preguntas en tu informe: ###
- **¬øQu√© es una API y c√≥mo puedes interactuar con ellas para obtener datos?**
Una API (Interfaz de Programaci√≥n de Aplicaciones) es un conjunto de reglas y protocolos que permite que diferentes aplicaciones se comuniquen entre s√≠. A trav√©s de una API, puedes enviar solicitudes y recibir respuestas de servicios externos. Para interactuar con una API y obtener datos, normalmente haces una solicitud HTTP, especificando el endpoint y los par√°metros necesarios. La respuesta suele estar en formato JSON, que luego puedes procesar en tu c√≥digo para extraer los datos que necesitas. Por ejemplo, si deseas obtener datos de Twitter, debes autenticarte con claves de API y hacer solicitudes a sus endpoints. Las respuestas se recibir√°n en formato JSON y puedes extraer, por ejemplo, los tweets que contienen ciertas palabras clave.


- **¬øCu√°l es la diferencia entre el web scraping y el uso de APIs?**
La principal diferencia entre el web scraping y el uso de APIs radica en la fuente de los datos y c√≥mo se accede a ellos:

    - APIs: Las APIs proporcionan una forma estructurada y estable de obtener datos, generalmente en formato JSON o XML. Los proveedores de servicios mantienen estas interfaces y te permiten obtener datos de manera controlada. Adem√°s, las APIs suelen tener documentaci√≥n y limitaciones sobre c√≥mo puedes usarlas.

    - Web Scraping: El web scraping consiste en extraer informaci√≥n directamente desde las p√°ginas web. No es una interfaz oficial, por lo que se debe extraer el contenido del HTML de una p√°gina. Esta t√©cnica es √∫til cuando no hay una API disponible o cuando necesitas acceder a datos que no est√°n disponibles a trav√©s de una API. Sin embargo, el scraping puede ser menos fiable, ya que depende de la estructura del sitio web, que puede cambiar sin previo aviso.

  
- **¬øC√≥mo puedes almacenar los datos adquiridos de una API o scraping en un archivo CSV?** Para almacenar los datos adquiridos de una API o mediante web scraping en un archivo CSV, debes seguir los siguientes pasos en Python:

1. Obtener los datos: Primero, haces la solicitud a la API o realizas el scraping de la p√°gina web.

2. Convertir los datos en formato adecuado: Los datos que recibes de una API o de scraping generalmente estar√°n en formato JSON o ser√°n un conjunto de objetos. Es necesario procesar esta informaci√≥n y convertirla en una estructura tabular adecuada para el formato CSV, como listas de diccionarios o listas de listas.

3. Escribir en un archivo CSV: Utilizas la biblioteca csv de Python para escribir los datos procesados en un archivo CSV.

### 2. Trabajo con Web Scraping (Ejemplo con BeautifulSoup): ###
- Paso 1: Instala BeautifulSoup: pip install beautifulsoup4 requests (en la terminal)
- Paso 2: Realiza scraping de un sitio web (por ejemplo, extraer los t√≠tulos de noticias de un sitio):

```python
from bs4 import BeautifulSoup # Importamos BeautifulSoup para analizar el contenido HTML
import requests # Importamos la librer√≠a requests para hacer peticiones HTTP
# Definimos la URL del sitio web que queremos scrape
url = 'https://www.bbc.com'
# Realizamos la solicitud HTTP al sitio web y obtenemos la respuesta
response = requests.get(url)
# Usamos BeautifulSoup para analizar el contenido de la respuesta en formato HTML
soup = BeautifulSoup(response.content, 'html.parser')
# Buscamos todas las etiquetas <h2> en el contenido HTML, que deber√≠an contener los titulares
headlines = soup.find_all('h2')
# Almacenamos los titulares en una lista
titles = []
# Recorremos cada uno de los elementos encontrados
for headline in headlines:
    text = headline.text.strip()
    if text: # Solo a√±adimos si no est√° vac√≠o
        titles.append(text)
# Imprimimos todos los titulares juntos, cada uno en una l√≠nea
print("\n".join(titles))
```

- Paso 3: Almacena los datos obtenidos en un archivo CSV:

```python
headlines_text = [headline.text for headline in headlines]
df = pd.DataFrame(headlines_text, columns=['Headline'])
df.to_csv('bbc_headlines.csv', index=False)
```
